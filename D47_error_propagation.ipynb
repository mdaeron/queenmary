# Error propagation for Δ47 measurements

The goal of this notebook is not to process real data, but rather to illustrate how Δ47 normalization errors behave, by allowing users to play around with different situations.

The functions used to process the data (anchors and unknowns), compute all errors, and generate plots, are defined in the accompanying file `D47errors.py`. To see what exactly is being done, look into that file.

Below, we define the number of anchor measurements and the composition of each anchor, as well as some unknown measurements. For each unknown we can then compare “inherent” and “normalization“ standard errors (noted `iSE`, `nSE`) and compute their total standard errors (`tSE`).

After changing the anchor/unknown data and/or the D47 reproducibility in the box below, select “Run” in the toolbar located near the top of your screen to recompute errore and update the plots. For some reason this doesn't work well the very first time, but the issue disappears when you click “Run” again.

```python
from D47errors import norm_error, compute_errors

# define anchor measurements
anchors = []
anchors += [{'name':'ETH-1', 'd47':+17, 'D47':0.258, 'N':12}]
anchors += [{'name':'ETH-2', 'd47':-11, 'D47':0.256, 'N':12}]
anchors += [{'name':'ETH-3', 'd47':+17, 'D47':0.691, 'N':4}]

# define unknown measurements
unknowns = []
unknowns += [{'name':'ETH-4',   'd47':-11, 'D47':0.513, 'N':4}]
unknowns += [{'name':'IAEA-C1', 'd47':+17, 'D47':0.360, 'N':4}]
unknowns += [{'name':'IAEA-C2', 'd47':0,   'D47':0.716, 'N':4}]
unknowns += [{'name':'MERCK',   'd47':-40, 'D47':0.573, 'N':4}]

# define session-wide D47 reproducibility
rD47 = 0.015

compute_errors(rD47, anchors, unknowns)
```
